# Agente de Búsqueda en Grafo aplicado para encontrar la ruta mínima para recorrer todas las iglesias del distrito San Miguel en el menor tiempo utilizando el algoritmo A* - Grupo 1
### 1. Obtención del Grafo Planar del Distrito de Miraflores con sus Iglesias

```python
from Agente import Agente
agent = Agente()
distrito = "San Miguel"

G = agent.descargarGrafoDistrito(distrito)
churches = agent.descargarIglesias(distrito)
church_nodes = agent.preprocesarGrafo(G, churches)

print(f"Se ha finalizado la preparación del grafo planar del Distrito de {distrito}.")
print(f"Número de Iglesis Identificadas: {len(churches)}")

print(f"Visualización del Grafo Planar de {distrito}:")
agent.visualizarGrafo(G, None, None)

print(f"Visualización del Grafo Planar de {distrito} con sus nodos Iglesias:")
agent.visualizarGrafo(G, church_nodes, None)
```

## 2. Estadisticas del Grafo

```python
print("Número de Ubicaciones (Nodos): ",G.number_of_nodes())
print("Número de Bordes (Aristas): ", G.number_of_edges())

import matplotlib.pyplot as plt
degrees = [degree for node, degree in G.degree()]
plt.hist(degrees, bins=max(degrees)-min(degrees)+1, edgecolor='black')
plt.title("Histograma del Grado de cada Ubicación")
plt.xlabel("Grado")
plt.ylabel("Frecuencia")
plt.xticks(range(min(degrees), max(degrees)+1))
plt.grid(True)
plt.show()

import networkx as nx
#shortest_path_length = nx.shortest_path_length(G.to_undirected(), source=1, target=5)
#print("Estimación Inicial de la Distancia Más Corta: ",shortest_path_length)

graph_radius = nx.radius(G.to_undirected())
print("Radio del Grafo: ", graph_radius)
graph_diameter = nx.diameter(G.to_undirected())
print("Diámetro del Grafo: ", graph_diameter)
```

# 2. Buscar la Ruta Optima del feligres para recorrer todas las Iglesias
## 2.1. Implementación con el Algoritmo BFS

```python
from SearchGraphBFS import GraphBfs
bfs = GraphBfs(G)
route = bfs.getRoute(church_nodes)
length = agent.calculate_route_length(G, route)
print(f"La distancia total de la ruta propuesta es de {length} metros.")

print("Visualizar la ruta encontrada:")
agent.visualizarGrafo(G, church_nodes, route)
```

## 2.2. Implementación con el Algoritmo DFS

```python
from SearchGraphDFS import GraphDfs
dfs = GraphDfs(G)
route = dfs.getRoute(church_nodes)
length = agent.calculate_route_length(G, route)
print(f"La distancia total de la ruta propuesta es de {length} metros.")

print("Visualizar la ruta encontrada:")
agent.visualizarGrafo(G, church_nodes, route)
```
